Описание проекта


TED (от англ. technology, education, design — «технологии, образование, дизайн») — некоммерческий фонд, который проводит популярные конференции. На них выступают специалисты из разных областей и читают лекции на актуальные социальные, культурные и научные темы. 
В разное время на TED-конференциях выступали математик Бенуа Мандельброт, теоретик искусственного интеллекта Марвин Минский, спортсменка Дана Ньяд и основатель Google Ларри Пейдж. В истории TED также были неоднозначные и даже скандальные выступления. Например, в 2010 году на конференции выступил Рэнди Пауэлл с рассказом о псевдонаучной «вихревой математике», а в 2014 году в конференции TEDMED участвовала Элизабет Холмс — основательница печально известного стартапа Theranos.


В этом проекте мы исследуем историю TED-конференций с помощью Tableau.

**Общий вывод по проекту:**

- Выступления чаще всего проходили в США, Канаде и Великобритании
- Наиболее популярны категории science и technology
- В 2020 и 2021 годах лидирующие позиции заняли категории society, global issues и technology
- Распределение популярных категорий в разных странах различается. Например в Канаде у категории science 24,28%, а в США у этой категории 18,64% 
- Чаще вызывают аплодисменты аудитории выступления категрии art, реже - global issues.
- Чаще вызывают смех аудитории выступления категрии art, реже - global issues.
- Зависимости между длительностью выступления и количеством просмотров практически нет.
- У авторов выступлений преобладают области деятельности writer и journalist.
- Oбычно  на одного автора приходится одно выступление.
- Состав тематик категории «Другие» в 2019–2021 годах изменялся следующим образом: категория perfomance уменьшилась в процентном отношении, в перед вышли категории humanity, personal growth, work. Однако по 2021 году оценки давать нельзя, слишком малое кол-во выступлений (3 выступления, по одному по каждой категории).
- Разделить конференции на категории не представляется возможным, по большинству конференций было несколько выступлений на разные тематики, порой долеких друг от друга по смыслу (к примеру в одной конференции выступлений 17-culture, 20-science, 15-technology). По названию конференции присвоить категорию также не представляется возможным.
